```python
import pandas as pd
```

# End Effector Positions

```python
end_effector_df1 = pd.read_csv("end_effector_positions_data_sample1.csv")
end_effector_df2 = pd.read_csv("end_effector_positions_data_sample2.csv")
end_effector_df3 = pd.read_csv("end_effector_positions_data_sample3.csv")
end_effector_df1.head()
```

```python
# Step 1: Determine the offsets for PairID for each subsequent dataframe
offset_df2 = end_effector_df1['PairID'].max()
offset_df3 = offset_df2 + end_effector_df2['PairID'].max()

# Step 2: Apply the offset to PairID in df2 and df3
end_effector_df2['PairID'] += offset_df2
end_effector_df3['PairID'] += offset_df3

# Step 3: Concatenate the dataframes
end_effector_combined_df = pd.concat([end_effector_df1, end_effector_df2, end_effector_df3], ignore_index=True)
```

```python
end_effector_combined_df
```

```python
end_effector_combined_df.to_csv("end_effector_positions_data.csv", index=False)
```

```python
test_df = pd.read_csv("end_effector_positions_data.csv")
test_df
```
